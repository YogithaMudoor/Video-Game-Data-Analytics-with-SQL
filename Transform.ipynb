**STEP 1 :TRANSFORM**
We need to transform the extracted data from VGchartz website into a cleaned data in this step. Following steps need to be in the transform phase
* Delete games without sales data

* Remove unwanted columns and rename the columns

* Add Release year column

* Convert Release date values to date format and Sales values to float format

* Write the transformed video games data into CSV file to be used for the Load process

The link for the extracted data is here - https://drive.google.com/file/d/1CWvgV4pLaMpJR2MmUnNdgbR7Gl4Sj5l5/view?usp=sharing

```python
import pandas as pd   #import pandas and datetime package
from datetime import datetime
```

**Read Video Games Data from CSV**

```python
# Read games data from CSV file into Dataframe
games_df1 = pd.read_csv('/content/drive/MyDrive/Extracted_video_games_data.csv')
```

```python
games_df1.head() #first 5 rows
```

**Delete games without sales data**

```python
games_df1.info() #info of all columns in dataframe
```

```python
games_df_withsales= games_df1.dropna( how='all',subset=['Total Sales','Total Shipped']) #drop rows where both total sales and total shipped both columns are null
games_df_withsales
```

**Delete unwanted columns**

Select only these columns :

'Pos', 'Game.1', 'Console', 'Publisher', 'Developer', 'VGChartz Score',
       'Critic Score', 'User Score', 'Total Shipped', 'Total Sales',
       'NA Sales', 'PAL Sales', 'Japan Sales', 'Other Sales', 'Release Date',
       'Genre'

```python
games_df_withsales.drop(['Game','Last Update'],inplace=True,axis=1)#deleting unwanted column
```

```python
games_df_withsales
```

**Rename columns**

```python
games_df_withsales.rename({'Game.1': 'game_name','Console':'console','Publisher':'publisher','Developer':'developer',\
                           'VGChartz Score':'vgchartz_score','Critic Score':'critic_score','User Score':'user_score',\
                           'Total Shipped':'total_shipped_in_millions','Total Sales':'global_sales_in_millions',\
                           'NA Sales':'na_sales_in_millions','PAL Sales':'pal_sales_in_millions','Japan Sales':'japan_sales_in_millions',\
                          'Other Sales':'other_sales_in_millions','Release Date':'release_date','Genre':'genre','Pos':'rank' }, axis=1, inplace=True)
```

**Add Release Year Column**

```python
games_df_cleaned=games_df_withsales.loc[games_df_withsales['release_date'].notnull()].copy()#copy rows where release date values are not null
```

```python

games_df_cleaned['release_date']=pd.to_datetime(games_df_withsales['release_date'])#convert release date column to datetime
```

```python
games_df_cleaned.info()
```

**Convert Release date values to Date format (yyyy-mm-dd)**

```python
games_df_cleaned['release_date']=games_df_cleaned['release_date'].dt.strftime('%Y-%m-%d') #convert release date column to string format(YYYY-MM-DD)

```

```python
games_df_cleaned['release_date']# display value of release date column
```

```python
games_df_cleaned['release_date']=games_df_cleaned['release_date'].astype('datetime64[ns]') #convert released date column to datetime
```

```python
games_df_cleaned.info()
```

```python
games_df_cleaned['release_year']=games_df_cleaned['release_date'].dt.year #Extract released year from release date column
```

```python
games_df_cleaned
```

```python
games_df_cleaned.info()
```

```python
games_df_cleaned.isnull().sum()
```

```python
def remove_m(row):#function to replace m from perticular column
  return row.replace("m","") 



```

```python
#lambda function used along with remove_m function to remove m from respective columns 
games_df_cleaned['global_sales_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['global_sales_in_millions'])),axis=1)
games_df_cleaned['na_sales_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['na_sales_in_millions'])),axis=1)
games_df_cleaned['pal_sales_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['pal_sales_in_millions'])),axis=1)
games_df_cleaned['japan_sales_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['japan_sales_in_millions'])),axis=1) 
games_df_cleaned['total_shipped_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['total_shipped_in_millions'])),axis=1)
games_df_cleaned['other_sales_in_millions']=games_df_cleaned.apply(lambda row:remove_m(str(row['other_sales_in_millions'])),axis=1)
```

**Convert Sales values to Float format**

```python

games_df_cleaned['global_sales_in_millions']=games_df_cleaned['global_sales_in_millions'].astype('float')
games_df_cleaned['na_sales_in_millions']=games_df_cleaned['na_sales_in_millions'].astype('float')  
games_df_cleaned['pal_sales_in_millions']=games_df_cleaned['pal_sales_in_millions'].astype('float') 
games_df_cleaned['japan_sales_in_millions']=games_df_cleaned['japan_sales_in_millions'].astype('float')     
games_df_cleaned['total_shipped_in_millions']=games_df_cleaned['total_shipped_in_millions'].astype('float') 
games_df_cleaned['other_sales_in_millions']=games_df_cleaned['other_sales_in_millions'].astype('float') 
```
